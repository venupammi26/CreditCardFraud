```python
#Outlier treatment

Q1 = data['Amount'].quantile(0.25)
Q3 = data['Amount'].quantile(0.75)
IQR = Q3 - Q1
data = data[(data['Amount'] >= Q1 - 1.5*IQR) & (data['Amount'] <= Q3 + 1.5*IQR)]
sn.boxplot(x=data['Amount'])
```

```python
data.shape
```

```python
#Checking balanced or imbalanced
data.Class.value_counts()
```

This is imbalnced data

# Treatment to imbalacnced data

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
#Oversampling

from imblearn.over_sampling import SMOTE

os = SMOTE(random_state = 0)
X2, y2 = os.fit_resample(X, y)
```

```python
X2=pd.DataFrame(X2,columns=X.columns)

frame = { 'Class': y2}
y2=pd.DataFrame(frame)
```

```python
y2['Class'].value_counts()
```

```python
data2=X2
data2['Class']=y2
data2.head()
```

```python
#Correlation
dfcor=data2.corr()
dfcor.style.background_gradient(cmap='coolwarm')
```

```python
dfcor['Class'].sort_values()
```

Here v15,V13,V22,V23,V8,V27,V25,V26 are lightly correlated with Target varaible.So reomve this varaiables

```python
data3=data2.drop(['V15','V13','V22','V23','V8','V27','V25','V26'],axis=1)

print(data3.shape)
data3.head()
```

```python
cr=data3.corr()
corr_variables=[]

for i in cr.columns:
    for j in cr.columns:
        if i!=j:
            if cr[i][j]>0.9:
                corr_variables.append([i,j,cr[i][j]])
          

if len(corr_variables)!=0:
    print('Highly Correlated indipendent variables are',corr_variables)
else:
    print('There is no correlated indipendent variables')    
```

```python
#Here V7 tightly correlated with V1 and V3 , V16 with V12 and V17 and V18 So remove V7 ,V16,V18 and V3
data4=data3.drop(['V7','V16','V3','V18'],axis=1)
print(data4.shape)
cr=data4.corr()
corr_variables=[]

for i in cr.columns:
    for j in cr.columns:
        if i!=j:
            if cr[i][j]>0.9:
                corr_variables.append([i,j,cr[i][j]])
          

if len(corr_variables)!=0:
    print('Highly Correlated indipendent variables are',corr_variables)
else:
    print('There are no correlated indipendent variables') 
```

```python
data4.head()
```

```python
X=data4.iloc[:,:-1]
y=data4.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

# Naive bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix,accuracy_score,roc_auc_score,roc_curve

nb=GaussianNB()
nb.fit(X_train,y_train)
y_pred=nb.predict(X_test)

print('Accuracy',accuracy_score(y_test,y_pred))
print('Confusion matrix:  \n', confusion_matrix(y_test,y_pred))
fpr, tpr, _ = roc_curve(y_test,  y_pred)
auc = roc_auc_score(y_test, y_pred)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)

RFC = RandomForestClassifier(n_estimators=300, random_state=0,bootstrap =True,criterion='entropy')
RFC.fit(X_train, y_train)
y_pred = FRFC.predict(X_test)

print('Accuracy',accuracy_score(y_test,y_pred))
print('Confusion matrix:  \n', confusion_matrix(y_test,y_pred))
fpr, tpr, _ = roc_curve(y_test,  y_pred)
auc = roc_auc_score(y_test, y_pred)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
